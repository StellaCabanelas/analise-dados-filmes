# Análise 1: Top 10 filmes com maior receita

```python
import matplotlib.pyplot as plt

# Selecionar as colunas que vamos usar
top_receita = filmes[['title', 'revenue']].sort_values(by='revenue', ascending=False).head(10)

# Criar o gráfico
plt.figure(figsize=(10,6))
plt.barh(top_receita['title'], top_receita['revenue'], color='skyblue')
plt.xlabel('Receita (em dólares)')
plt.ylabel('Filme')
plt.title('Top 10 Filmes com Maior Receita')
plt.gca().invert_yaxis()  # inverte para o maior ficar no topo
plt.show()
```

🎬 Observa-se que "Avatar" lidera o ranking de receita, seguido por grandes produções como "Star Wars: The Force Awakens" e "Titanic".

# Análise 2: Quantidade de filmes lançados por ano

```python
# 1) extrair ano da coluna 'release_date' (se já tiver 'release_year', pule essa linha)
filmes['release_year'] = pd.to_datetime(filmes['release_date'], errors='coerce').dt.year

# 2) agrupar e ordenar
filmes_por_ano = filmes.groupby('release_year').size().reset_index(name='quantidade').dropna().sort_values('release_year')

# 3) gráfico de tendência
plt.figure(figsize=(12,6))
plt.plot(filmes_por_ano['release_year'], filmes_por_ano['quantidade'], marker='o')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Quantidade de Filmes')
plt.title('Quantidade de Filmes Lançados por Ano')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Décadas iniciais: 
Número reduzido de lançamentos, reflexo de uma indústria ainda em desenvolvimento e com limitações tecnológicas e de distribuição.

### Crescimento gradual: 
A partir dos anos 1980/1990, há um aumento consistente, impulsionado pela popularização do cinema, avanços tecnológicos e expansão do mercado global.

### Picos e quedas: 
Alguns anos apresentam quedas bruscas — possivelmente relacionadas a crises econômicas, greves na indústria ou eventos globais que afetaram as produções.

### Últimos anos: 
Tendência de alta até certo ponto recente, podendo depois apresentar uma leve redução (ex.: impactos da pandemia na indústria a partir de 2020).

# Análise 3: Receita média por gênero

```python
import ast

# 1) Garantir que 'revenue' esteja numérico
filmes['revenue_clean'] = pd.to_numeric(filmes['revenue'], errors='coerce')

# 2) Converter a coluna 'genres' (texto no formato de lista de dicionários) em listas reais
def extrair_generos(generos_str):
    try:
        lista = ast.literal_eval(generos_str)  # converte string para lista Python
        return [g['name'] for g in lista] if isinstance(lista, list) else []
    except (ValueError, SyntaxError):
        return []

filmes['lista_generos'] = filmes['genres'].apply(extrair_generos)

# 3) "Explodir" os gêneros — cada gênero vira uma linha separada
filmes_explodido = filmes.explode('lista_generos')

# 4) Calcular receita média por gênero
receita_por_genero = filmes_explodido.groupby('lista_generos')['revenue_clean'].mean().reset_index()
receita_por_genero = receita_por_genero.dropna().sort_values(by='revenue_clean', ascending=False)

# 5) Criar gráfico
plt.figure(figsize=(10,6))
plt.barh(receita_por_genero['lista_generos'], receita_por_genero['revenue_clean'])
plt.xlabel('Receita Média (em dólares)')
plt.ylabel('Gênero')
plt.title('Receita Média por Gênero')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

### Gêneros mais lucrativos:
Animation, Adventure e Fantasy lideram, com receita média próxima ou acima de 200 milhões de dólares.
Isso sugere que filmes com forte apelo visual e narrativas imaginativas atraem grandes públicos e investimentos, possivelmente com maior bilheteria global e merchandising.

### Público familiar como força motriz:
Animation e Family aparecem entre os primeiros, indicando que produções voltadas para todas as idades têm potencial de receita muito alto, provavelmente por atingirem múltiplas faixas etárias e mercados internacionais.

### Gêneros de ação e ficção científica:
Science Fiction e Action também aparecem bem posicionados, reforçando que produções com efeitos especiais, aventuras e alta intensidade têm bom retorno financeiro.

### Receita média mais baixa:
Documentary, Foreign e TV Movie têm as menores receitas médias. Isso pode estar ligado a públicos mais nichados, distribuição limitada e orçamentos mais baixos.

### Curiosidade:
Alguns gêneros como War e Thriller apresentam receita intermediária, o que pode significar que, apesar de serem populares, sua performance depende muito de títulos específicos e não do gênero como um todo.

## Implicações Estratégicas para Produção e Marketing

Com base nos dados de receita média por gênero, podemos pensar em três linhas de ação estratégicas:  


### 1. Priorização de gêneros de alto retorno  
- **Animation**, **Adventure** e **Fantasy** se destacam como apostas seguras para grandes produções.  
- Para estúdios e investidores, isso significa que orçamentos maiores e campanhas de marketing massivas tendem a ter mais retorno nesses segmentos, especialmente com distribuição global.  


### 2. Expansão de apelo internacional  
- Gêneros como **Animation**, **Science Fiction** e **Action** têm narrativas universais e dependem mais de elementos visuais que de nuances culturais ou linguísticas, o que facilita a exportação e aumenta bilheterias no exterior.  
- **Estratégia:** investir em dublagens de alta qualidade, pré-lançamentos em mercados estratégicos (China, Índia, América Latina) e parcerias com plataformas de streaming globais.  


### 3. Segmentação e nicho para gêneros de baixa receita média  
- **Documentary**, **Foreign** e **TV Movie** ainda podem ser rentáveis quando se trabalha com baixo custo de produção e estratégias digitais de distribuição.  
- Aqui, campanhas segmentadas, festivais de cinema e parcerias com plataformas especializadas podem maximizar retorno sem exigir grandes investimentos. 


## Tendências Históricas e Possibilidades de Projeção

A partir da evolução de receita média ao longo dos anos, é possível extrair três pontos estratégicos:  

### 1. Identificação de ciclos de alta e baixa  
- Algumas categorias apresentam **picos de receita** em períodos específicos, geralmente associados a grandes franquias ou fenômenos culturais.  
- **Exemplo:** um aumento repentino em **Fantasy** ou **Science Fiction** pode estar ligado ao lançamento de sagas populares.  
- **Aplicação:** planejar lançamentos para coincidir com períodos de alta demanda no gênero.  

### 2. Projeção de tendências futuras  
- Ao observar o crescimento constante em gêneros como **Animation** e **Adventure**, há indícios de que o público está cada vez mais receptivo a narrativas escapistas e visuais.  
- **Estratégia:** alinhar a pipeline de produção para atender a essa demanda, priorizando roteiros e produções com forte apelo visual e potencial para franquias.  

### 3. Potencial de resgate de gêneros em baixa  
- Gêneros como **Musical** ou **Western**, que apresentam queda de receita, ainda podem ter sucesso com abordagens modernas ou fusão com outros gêneros.  
- **Exemplo:** um musical com elementos de **Comedy** ou **Drama** pode alcançar novos públicos.  
- **Aplicação:** apostar em inovações narrativas para revitalizar segmentos menos explorados.  

```python
# Converter budget e revenue para numérico (se já não estiver)
filmes["budget"] = pd.to_numeric(filmes["budget"], errors='coerce')
filmes["revenue"] = pd.to_numeric(filmes["revenue"], errors='coerce')

# Remover filmes com budget 0 para evitar divisão por zero e dados ruins
filmes = filmes[filmes["budget"] > 0].copy()  # <-- aqui o .copy() para evitar warning

# Calcular ROI
filmes["ROI"] = (filmes["revenue"] - filmes["budget"]) / filmes["budget"]

# Extrair gênero principal da coluna 'genres' (que é string JSON)
def extrair_genero_principal(genres_str):
    try:
        genres = ast.literal_eval(genres_str)  # Converte string para lista de dicts
        if len(genres) > 0:
            return genres[0]["name"]
        else:
            return None
    except:
        return None

filmes["genero_principal"] = filmes["genres"].apply(extrair_genero_principal)

# Remover linhas sem gênero principal
filmes = filmes.dropna(subset=["genero_principal"]).copy()  # Também .copy() aqui para evitar warning

# Scatter Plot Budget x Revenue 
plt.figure(figsize=(8,6))
plt.scatter(filmes["budget"], filmes["revenue"], alpha=0.6, c="skyblue", edgecolor="k")
plt.plot([filmes["budget"].min(), filmes["budget"].max()],
         [filmes["budget"].min(), filmes["budget"].max()],
         color="red", linestyle="--", label="Ponto de equilíbrio (Orçamento = Receita)")
plt.xlabel("Orçamento")
plt.ylabel("Receita")
plt.title("Orçamento vs Receita (Linha de equilíbrio em vermelho)")
plt.legend()
plt.show()
```
